ANS- 
Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in various formats. It provides a wide range of functions and tools for generating high-quality plots and charts, making it one of the most widely used plotting libraries in the data science and scientific communities.

Key features and reasons for using Matplotlib include:

Easy to Use: Matplotlib provides a simple interface for creating plots and charts, making it accessible to beginners while still offering advanced customization options for more experienced users.

Publication-Quality Graphics: It allows users to create professional-quality plots for publications, presentations, and reports with fine-grained control over elements like fonts, colors, and labels.

Customizability: Matplotlib offers extensive customization options, allowing users to modify virtually every aspect of the plot to suit their specific needs.

Wide Range of Plot Types: Matplotlib supports a variety of plot types, making it versatile for data visualization across different domains and applications.

Integration with NumPy: Matplotlib integrates seamlessly with NumPy, a fundamental library for numerical computing in Python, making it convenient to visualize data stored in NumPy arrays.

The pyplot module of Matplotlib provides a simple and state-based interface for creating plots. Here are five commonly used plots that can be created using the pyplot module:

Line Plot: A basic plot that displays data points connected by straight line segments, typically used to show trends or patterns over continuous intervals.

Scatter Plot: This plot displays individual data points as separate dots, useful for visualizing the relationship between two variables and detecting clusters or outliers.

Bar Plot: A plot that represents categorical data with rectangular bars, where the height or length of each bar corresponds to the data value for the category.

Histogram: Used to visualize the distribution of a continuous variable by dividing it into bins and displaying the frequency of values falling into each bin.

Pie Chart: A circular plot divided into sectors, where the area of each sector is proportional to the data value it represents. Pie charts are suitable for displaying the composition of a whole as parts.

Matplotlib's capabilities extend beyond these five types, and it can handle other types of plots like area plots, box plots, heatmap plots, and more. Its versatility and user-friendly nature make it a preferred choice for data visualization in Python

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python

ANS-
import numpy as np
import matplotlib.pyplot as plt

# Seed the random number generator for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create the scatter plot
plt.scatter(x, y, marker='o', color='blue', label='Data Points')

# Add title and labels for x and y axes
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

ANS--The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to divide the figure into a grid of rows and columns and specify the position of each subplot using row and column indices. This is useful when you want to compare multiple plots or visualize different aspects of your data side by side.

To create four line plots using the subplot() function, you can use the provided data for each line and specify the position of each subplot in a 2x2 grid.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Data for Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')

# Add a common legend and adjust layout
plt.legend()
plt.tight_layout()

# Display the subplots
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

ANS-A bar plot, also known as a bar chart, is a data visualization that represents categorical data with rectangular bars. The length or height of each bar corresponds to the value of the data it represents. Bar plots are used to compare different categories and display the distribution of data within each category.

Bar plots are useful for the following purposes:

Comparison: They allow easy comparison of data between different categories, making it simple to identify which category has the highest or lowest value.

Categorical Data Visualization: Bar plots are particularly effective when dealing with categorical data, where each category has distinct labels.

Visualizing Discrete Data: Bar plots are suitable for visualizing data with discrete values, such as counts or frequencies of occurrences.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit (Millions)')
plt.title('Company Profits')
plt.grid(axis='y')
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

ANS -A box plot, also known as a box-and-whisker plot, is a data visualization that provides a summary of the distribution of a dataset. It displays the five-number summary of the data, which includes the minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. The box in the plot represents the interquartile range (IQR) of the data, and the "whiskers" extend from the box to indicate the range of the data beyond the quartiles.

The main components of a box plot are as follows:

Minimum: The smallest value in the dataset, excluding any outliers.

Maximum: The largest value in the dataset, excluding any outliers.

Q1 (First Quartile): The 25th percentile of the data, separating the lowest 25% of the data from the rest.

Q2 (Median): The 50th percentile of the data, representing the middle value when the data is sorted.

Q3 (Third Quartile): The 75th percentile of the data, separating the lowest 75% of the data from the highest 25%.

Interquartile Range (IQR): The range between Q1 and Q3 (IQR = Q3 - Q1).

Outliers: Individual data points that are significantly different from the rest of the data.

Box plots are used to:

Identify Skewness: Box plots allow us to quickly see if the data is symmetrically distributed or skewed to one side.

Detect Outliers: Outliers are shown as individual points beyond the whiskers, making them easy to identify.

Compare Distributions: Box plots are useful for comparing the distribution of multiple datasets side by side.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data for plotting
data_to_plot = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data_to_plot, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot Comparison')
plt.grid(axis='y')
plt.show()
```
